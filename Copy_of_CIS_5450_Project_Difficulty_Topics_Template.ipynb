### Implementation & Rationale


**How we implemented it (where in code):**
- Ran the cleaning pipeline via `scripts/pipeline/run_cleaning.py` (or the same logic called from the notebook).
- Core pipeline logic lives in `src/cleaning/hotel_world_clean.py`:
  - `clean_hotels()` engineers features like numeric star ratings, parsed lat/long, and text-derived counts.
  - `clean_world_cities()` standardizes city keys and dedupes to one row per `(country_iso2, city_join_key)`.
  - `join_hotels_world_cities()` merges hotels ↔ cities and adds `city_population`.

**Why this was hard / why we had to do it:**
- **Schema + environment friction:** hotel CSV headers/encodings vary across exports (and Kaggle vs local), so we had to normalize inputs to make results reproducible.
- **Messy text → usable signal:** key info lives in unstructured fields (`Attractions`, `HotelFacilities`), so we engineered compact numeric features instead of feeding raw strings.
- **Non-trivial joins:** matching hotels to world-cities needs cleaned join keys and careful deduping to avoid incorrect matches.

We had to do this so modeling isn’t built on inconsistent columns, broken joins, or unusable text—otherwise results would be brittle and misleading.

### Implementation & Rationale
To further understand which factors most strongly influence hotel ratings, feature importance techniques were applied using a regression or tree-based model. This approach allows us to quantify the relative contribution of each feature, such as facility count or presence of attractions information, in predicting hotel ratings. The results reinforce earlier findings, showing that facility-related variables have a significantly higher impact on ratings compared to features like attractions information, which contribute little predictive value. This adds a more rigorous, model-based perspective to the exploratory analysis and helps validate which variables are truly meaningful for prediction.

#### How did you implement this topic and why did you use it?

We implemented feature importance to better understand which variables had the strongest influence on predicting hotel star ratings, instead of only relying on visual patterns or correlations.

In our project, this was especially useful because we had multiple possible predictors, such as facility count and attractions information. By comparing their importance scores, we could see which variables were actually useful for the model and which contributed very little. This helped confirm our earlier exploratory findings: facility-related features were more meaningful predictors of hotel rating, while features like attractions information appeared to reflect listing detail rather than hotel quality. It also helped to improve model interpretability, as Hotel ratings are influenced by many overlapping factors, so feature importance helped us move beyond simple observation and identify which features carried the most predictive value. This made our analysis more rigorous and helped justify why certain variables were emphasized in the final model.
