### Implementation & Rationale


**How we implemented it (where in code):**
- Ran the cleaning pipeline via `scripts/pipeline/run_cleaning.py` (or the same logic called from the notebook).
- Core pipeline logic lives in `src/cleaning/hotel_world_clean.py`:
  - `clean_hotels()` engineers features like numeric star ratings, parsed lat/long, and text-derived counts.
  - `clean_world_cities()` standardizes city keys and dedupes to one row per `(country_iso2, city_join_key)`.
  - `join_hotels_world_cities()` merges hotels ↔ cities and adds `city_population`.

**Why this was hard / why we had to do it:**
- **Schema + environment friction:** hotel CSV headers/encodings vary across exports (and Kaggle vs local), so we had to normalize inputs to make results reproducible.
- **Messy text → usable signal:** key info lives in unstructured fields (`Attractions`, `HotelFacilities`), so we engineered compact numeric features instead of feeding raw strings.
- **Non-trivial joins:** matching hotels to world-cities needs cleaned join keys and careful deduping to avoid incorrect matches.

We had to do this so modeling isn’t built on inconsistent columns, broken joins, or unusable text—otherwise results would be brittle and misleading.
